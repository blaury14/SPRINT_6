## Preparación de los datos

### Renombrar columnas:

```python
# Renombrar las columnas a minúsculas
data.columns = data.columns.str.lower()
```

### Conversión de tipos de datos:

```python
# Tratando NaNs en 'year_of_release'
data['year_of_release'] = data['year_of_release'].fillna(0).astype(int)

# Convirtiendo 'user_score' a numérico, manejando 'tbd' como NaN
data['user_score'] = pd.to_numeric(data['user_score'], errors='coerce')
```

### Tratamiento de valores ausentes:

```python
# Tratar valores ausentes en la columna 'name' y 'genre'
data['name'].fillna('Unknown', inplace=True)
data['genre'].fillna('Unknown', inplace=True)
# Tratar valores ausentes en la columna 'year_of_release' utilizando la mediana
median_year = data['year_of_release'].median()
data['year_of_release'].fillna(median_year, inplace=True)

# Tratar valores ausentes en la columna 'critic_score' y 'user_score' utilizando la media
mean_critic_score = data['critic_score'].mean()
data['critic_score'].fillna(mean_critic_score, inplace=True)

mean_user_score = data['user_score'].mean()
data['user_score'].fillna(mean_user_score, inplace=True)

# Tratar valores ausentes en la columna 'rating' utilizando 'Unknown'
data['rating'].fillna('Unknown', inplace=True)
```

### Cálculo de ventas totales:

```python
# Calcular las ventas totales y agregarlas en una nueva columna 'total_sales'
data['total_sales'] = data['na_sales'] + data['eu_sales'] + data['jp_sales'] + data['other_sales']
```

Además, las consolas de generaciones anteriores aún tienen un desempeño comercial considerable, posiblemente debido a una base instalada sólida y una amplia selección de juegos. Por otro lado, las consolas más recientes como la PS4 y la Xbox One están en proceso de acumular ventas, mientras que las más antiguas, como la NES y la SNES, tienen una presencia menor en el mercado actual. Además, se eligió la Xbox 360 como objeto de estudio para analizar la influencia de las reseñas en las ventas debido a su relevancia durante el período de los datos. Lanzada en 2005, la Xbox 360 ha sido una de las consolas líderes en el mercado durante ese tiempo, lo que la convierte en una opción ideal para comprender cómo la percepción de calidad afecta el éxito comercial en la industria moderna de los videojuegos.

### Deviacion Estandar y Viarianza

```python
# Desviación estándar de las ventas totales
std_total_sales = data['total_sales'].std()
print(f"La desviación estándar de las ventas totales por juego es: {std_total_sales}")

# Varianza de las ventas totales
variance_total_sales = data['total_sales'].var()
print(f"La varianza de las ventas totales por juego es: {variance_total_sales}")
```

**Conclusion** 
* La desviación estándar de alrededor de 1.55 revela que las ventas de juegos tienen una variabilidad moderada en relación con la media. Esto sugiere que la mayoría de las ventas de juegos se encuentran dentro de un rango relativamente cercano al promedio, aunque algunos juegos destacan significativamente por encima o por debajo de esta medida. 
* La varianza, calculada como el cuadrado de la desviación estándar (aproximadamente 2.4 en este caso), confirma esta dispersión. ## Prueba las siguientes hipótesis:

### Prueba de Hipótesis 1 - Comparación de plataformas Xbox One y PC.

* Hipótesis nula (H0): La calificación promedio de los usuarios para las plataformas Xbox One (XOne) y PC es la misma. 
* Hipótesis alternativa (H1): La calificación promedio de los usuarios para las plataformas XOne y PC no es la misma.

Para probar esta hipótesis, usaríamos una prueba T de muestras independientes, ya que estamos comparando las medias de dos muestras independientes.

```python
# Definimos el valor de alfa
alpha = 0.05

# Preparación de datos: asegurándose de que 'user_score' no tiene 'tbd' y convertir a numérico
data['user_score'] = pd.to_numeric(data['user_score'], errors='coerce')

# Filtrando los datos para cada plataforma
xone_user_scores = data[(data['platform'] == 'XOne') & (data['user_score'].notna())]['user_score']
pc_user_scores = data[(data['platform'] == 'PC') & (data['user_score'].notna())]['user_score']

# Realizando la prueba T para la primera hipótesis
t_stat, p_value = stats.ttest_ind(xone_user_scores, pc_user_scores)

print('Prueba de Hipótesis 1 - Comparación de plataformas Xbox One y PC:')
print('T-Statistic:', t_stat)
print('P-Value:', p_value)

if p_value < alpha:
    print("Rechazamos la hipótesis nula: Las calificaciones promedio de los usuarios para las plataformas Xbox One y PC son diferentes.")
else:
    print("No rechazamos la hipótesis nula: No hay diferencia significativa en las calificaciones promedio de los usuarios para las plataformas Xbox One y PC.")
```

**Conclusión de la Prueba de Hipótesis 1:**
Se rechazó la idea de que las calificaciones promedio de los usuarios para Xbox One y PC son iguales. Con un p-value extremadamente bajo (menor que 0.0001), hay evidencia sólida de que hay diferencias significativas en las calificaciones de usuario promedio entre estas plataformas. Esto sugiere que los usuarios perciben y califican los juegos de manera diferente en Xbox One y PC.

### Prueba de Hipótesis 2 - Comparación de géneros Acción y Deportes.

* Hipótesis nula (H0): La calificación promedio de los usuarios para los géneros de Acción y Deportes es la misma.
* Hipótesis alternativa (H1): La calificación promedio de los usuarios para los géneros de Acción y Deportes no es la misma.

Al igual que con la primera hipótesis, una prueba T de muestras independientes es adecuada para esta comparación.

```python
# Filtrando los datos para cada género
action_user_scores = data[(data['genre'] == 'Action') & (data['user_score'].notna())]['user_score']
sports_user_scores = data[(data['genre'] == 'Sports') & (data['user_score'].notna())]['user_score']

# Realizando la prueba T para la segunda hipótesis
t_stat, p_value = stats.ttest_ind(action_user_scores, sports_user_scores)

print('\nPrueba de Hipótesis 2 - Comparación de géneros Acción y Deportes:')
print('T-Statistic:', t_stat)
print('P-Value:', p_value)

if p_value < alpha:
    print("Rechazamos la hipótesis nula: Las calificaciones promedio de los usuarios para los géneros de Acción y Deportes son diferentes.")
else:
    print("No rechazamos la hipótesis nula: No hay diferencia significativa en las calificaciones promedio de los usuarios para los géneros de Acción y Deportes.")
```

**Conclusión de la Prueba de Hipótesis 2:**
No hay suficiente evidencia estadística para decir que hay una diferencia en las calificaciones promedio de los usuarios entre los géneros de Acción y Deportes. El p-value obtenido fue mayor que el umbral alfa de 0.05, lo que sugiere que cualquier diferencia observada podría ser aleatoria. Esto implica que, en general, los juegos de Acción y Deportes son valorados de manera similar por los usuarios.
